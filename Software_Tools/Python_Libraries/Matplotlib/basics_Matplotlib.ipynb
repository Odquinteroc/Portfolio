```python
#Task1
import matplotlib.pyplot as plt
import numpy as np


from matplotlib import style
#style.use("bmh")

#Task2
x = np.arange(1,101) #Make a list (array) of values from 1 to 100) remmember that 101 is not contained in the list
y1 = np.array(x**2) # takes each value from X list and returns its square and appends it to the y1 array
y2 = np.array(x**3) # takes each value from X list and returns its cubic value and appends it to the y2 array

font1 = {'family':'serif','color':'blue','size':10} 
font2 = {'family':'serif','color':'red','size':8} # make a variable for change the size, color  and the letter shape.

#task 3
plt.subplot(2, 2, 1)  # the first number means 2 rows of plots, the secund number means 2 columns, the third number (1) means the index positions.

plt.plot(x, y1, label="X^2", color = "y", linestyle = "dashed" ) # plot the variable x and y1 , add a label, change the defualt color to yellow, and change the lines style to dashed
plt.title("X vs Y1", fontdict = font1) # Add a label to the title of the plot
plt.xlabel("X",  fontdict = font2,) # Add a label to 'X' axis  of the plot
plt.ylabel("Y1",  fontdict = font2)
plt.grid(axis= 'x',color = "red", linestyle = '--') # add a grid dashed line on x- axis changenging the color 
plt.legend( loc='lower right') #allow add label on the plot,also change the legend position to lower right

#task4
plt.subplot(2, 2, 2) # the first number means 2 rows of plots, the secund number means 2 columns, the third number (2) means the index positions.
plt.plot(x,y2,'bo' , label="X^3")
plt.axis([0, 100, 0, 10000]) # the .axis function allow you to control the scaling and limits of the x and y axes ([xmin, xmax, ymin, ymax])
plt.title("Scatter plots", fontdict = font1)
plt.xlabel("X", fontdict = font2) 
plt.ylabel("y2",fontdict = font2)
plt.grid( axis='y', color = "m", linestyle = ":") # add a grid dotted on y - axis  changenging the color 
plt.legend(loc='lower left')

#Task 5
Z = np.random.randn(1000), # the numpy library with the function random.randn() returns and sample of normal distribution with 1000 random samples
plt.subplot(2, 2, 3)
plt.hist(Z, bins =50, edgecolor = "black", label = "N" ) 
plt.title("Normal Distribution", fontdict = font1)
plt.xlabel("X", fontdict = font2) 
plt.ylabel("f(x)",fontdict = font2)
plt.legend()

#plt.subplots_adjust(hspace=1)

plt.tight_layout() #  the function .tight_layout adjust the padding between each plot

#Task6 is implicit in the Task number 3 and 4 because I did both plots using the function plt.subplot

#Task 8
# in each plot I did some customizations like legend positioning, themes, colors, grid lines

sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

# Plot
plt.subplot(2, 2, 4)
#plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=140,textprops={'fontsize':5})
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example', fontdict = font1)


#Task8
plt.savefig("MyfistSavings", bbox_inches = "tight", dpi=300) # the function .savefig create a new image file without extra white spaces around the plot

plt.show() # The functions .show allow you display the plot created by matplotlib because it does not display automatically


```
